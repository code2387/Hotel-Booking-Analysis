```python
# Convert index to a list
a = df['previous_cancellations'].value_counts().index.to_list()
# Sort the list by ascending order
a.sort()
a
```

```python
# Finding the current cancellation rate according to their previous cancellation
b = []
for ccltime in a:
    b.append(round(df[df['previous_cancellations']==ccltime]['is_canceled'].mean() * 100 , 2))
print(b)
```

```python
# Creating a dictionary with keys as previous_cancellation and values as current_cancellation_rate
cancellation_rate = {'previous_cancellation': a, 'current_cancellation_rate': b}
cancellation_rate_data = pd.DataFrame(cancellation_rate)
cancellation_rate_data
```

```python
# Select single, couple, multiple adults and family
df_not_canceled = df[df['is_canceled'] == 0]

# Select single, couple, multiple adults and family
single   = df_not_canceled[(df_not_canceled.adults == 1) & (df_not_canceled.children == 0) & (df_not_canceled.babies == 0)]
couple   = df_not_canceled[(df_not_canceled.adults == 2) & (df_not_canceled.children == 0) & (df_not_canceled.babies == 0)]
family   = df_not_canceled[df_not_canceled.adults + df_not_canceled.children + df_not_canceled.babies > 2]
```

```python
# Selecting three colums from DataFrame which are: "is_canceled", "reserved_room_type", "assigned_room_type"
df[['is_canceled','reserved_room_type','assigned_room_type']]
```

```python
def rrt_art_same(df):
  '''
  This function returns whether the customer get their preferred reserved room type or not.
  '''
  if df['reserved_room_type'] == df['assigned_room_type']:
      return 'True'
  else:
      return 'False'
```

```python
df['rrt_art_same'] = df.apply(rrt_art_same,axis=1)

df[['is_canceled','reserved_room_type','assigned_room_type','rrt_art_same']]
```

```python
p1 = df['reserved_room_type'].value_counts().reset_index()
p2 = df['assigned_room_type'].value_counts().reset_index()

p1.columns = ['room_type', 'reserved_count']
p2.columns = ['room_type', 'assigned_count']

merged_RJ_df = pd.merge(p1, p2, how='right', on='room_type')

merged_RJ_df.fillna(0, inplace=True)

merged_RJ_df = merged_RJ_df.set_index('room_type')

print(merged_RJ_df)
```

```python
# Plotting
plt.figure(figsize=(12, 6))
plt.plot(merged_RJ_df.index, merged_RJ_df['reserved_count'], marker='o', label='Reserved Room Count', color='blue')
plt.plot(merged_RJ_df.index, merged_RJ_df['assigned_count'], marker='o', label='Assigned Room Count', color='green')

plt.title('Comparison between Reserved and Assigned Room Counts')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.xticks(rotation=45)  
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
